```python
# columns to drop --> link, property_id
df.drop(columns=['link', 'property_id'], inplace=True)
```

```python
df.head()
```

```python
# rename the colums area to price_per_sqft
df.rename(columns={'area': 'price_per_sqft'}, inplace=True)
```

```python
df.head()
```

```python
# work on socity column
df['society'].value_counts(ascending=False)
```

```python
# thier is a proble same socity with differnt spaicial character is count as two diffrent objects
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
# Price
df['price'].value_counts()
```

```python
 #i want  to convert all values in crore and also remove Lac behind the number

df = df[df['price'] != 'Price on Request']
```

```python
def convert_to_crore(value):
    if pd.isna(value):  # Handle NaN values
        return np.nan

    value = str(value).strip()
    if value == "" or not any(x in value for x in ["Lac", "Crore"]):
        return np.nan   # Skip invalid entries

    try:
        if "Lac" in value:
            return float(value.replace("Lac", "").strip()) / 100
        elif "Crore" in value:
            return float(value.replace("Crore", "").strip())
    except ValueError:
        return np.nan   # If still invalid, return NaN

# Apply function
df["Price_in_Crore"] = df["price"].apply(convert_to_crore)
```

```python
#drop price cloumn
df.drop(columns=['price'], inplace=True)
```

```python
df.head()
```

```python
# prices_per_sqft
df['price_per_sqft'].value_counts()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip()
df['price_per_sqft'] = pd.to_numeric(df['price_per_sqft'], errors='coerce')
```

```python
df.head()
```

```python
#bedrooms
df = df[~df['bedRoom'].isnull()]
```

```python
df['bedRoom'] = df['bedRoom'].astype(str).str.split(' ').str.get(0)
df = df[df['bedRoom'].str.isnumeric()]
df['bedRoom'] = df['bedRoom'].astype(int)
```

```python
df.head()
```

```python
# similarlly do for balcony and bathrooms
df['balcony'].isnull().sum()
```

```python
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')

df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype('int')
```

```python
df.head()
```

```python
# additional rooms

df['additionalRoom'].fillna('not available',inplace=True)
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df.head()
```

```python
#floor number

df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
```

```python
#facing

df['facing'].fillna('NA',inplace=True)
```

```python
df.insert(loc=4,column='area',value=round((df['Price_in_Crore']*10000000)/df['price_per_sqft']))
```

```python
df.head()
```

```python
df.insert(loc=1,column='property_type',value='flat')
```
